### 1.A description of the problem and a discussion of the background


São Paulo is the most populous city in Brazil and South America, being one of the most important cities in the world for gastronomic tourism. Its cuisine is so diverse that it pleases the street food lovers as much as the most demanding palates. 

Also,São Paulo is the city that consumes the most pizzas in Brazil (720 per minute) and is the second largest consumer in the world (losing only to New York).

As the figures tell, Sao Paulo attracts many to start their business in the Pizza industry. Before they take action, they need to find out where they would open it. What would they consider when selecting a location? By exploring the regional characteristics of restaurants, the goal is to figure out whether the neighborhood of restaurants is an essential factor for the success of a Pizza restaurant.
